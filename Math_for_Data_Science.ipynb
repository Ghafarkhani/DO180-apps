```python
# Vista

import matplotlib.pyplot as plt

x = []
y = []
t = 0

while t <= 5:
  h = (-4.9 * (t**2)) + (22.3 * t) + 2.1
  h = round(h, 2)
  x.append(t)
  y.append(h)
  t = t + 0.1
  t = round(t, 1)

plt.plot(x,
         y,
         color='green',
         linestyle='dashed',
         linewidth=2,
         marker='.',
         markerfacecolor='blue',
         markersize=2)
```

```python
from sympy import *
from sympy.plotting import plot3d

x, y = symbols('x y')
f1 = 3 * x
f2 = x**2 + 1
f3 = 2*x + 3*y

plot(f1, f2)
plot3d(f3)
```

```python
summation_1 = sum(2*i for i in range(1,6))
print("summation_1 = ", summation_1)

x = [1, 2, 3, 4]
n = len(x)
summation_2 = sum(10*x[i] for i in range(0,n))
print("summation_2 = ", summation_2)
```

```python
from sympy import *

i, n = symbols('i n')
summation = Sum(2*i, (i,1,n))
up_to_5 = summation.subs(n,5)

print("up_to_5 = ", up_to_5.doit())
```
